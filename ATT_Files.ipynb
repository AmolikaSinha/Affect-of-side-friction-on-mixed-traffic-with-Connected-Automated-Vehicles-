```python
df_ATT = pd.read_csv(r'C:\Users\z5130037\OneDrive - UNSW\Honours student Sai\Danial\Original_data\3X3\ATT_summary.csv', delimiter = ',', index_col=False)#, names = ['Time','Link','Lane', 'VehNo','VehType', 'Line', 'DesSpeed'], dtype = {'Time' : float , 'Link' : float, 'Lane' : float, 'VehNo' : float, 'VehType' : float,'Line' : float, 'DesSpeed' : float }, skiprows =1, index_col=False)

lengthATT = len(df_ATT)
#print(lengthATT)
df_ATT
```

```python
#DATA WITH AUTONOMOUS BUSES
TravelTime_AB = np.zeros((int(lengthATT/2),2))
DelayAverage_AB = np.zeros((int(lengthATT/2),2))
AverageSpeed_AB = np.zeros((int(lengthATT/2),2))
LatentDemand_AB = np.zeros((int(lengthATT/2),2))
EffectiveDelay_AB = np.zeros((int(lengthATT/2),2))

#print(TravelTime_AB[0,1])
#print(df_ATT.iloc[1,lengthATT])
```

```python
#DATA WITH AUTONOMOUS BUSES
for a in range(0,lengthATT):

    if (a%2) != 0:
        
        TravelTime_AB[int(a/2),0] = df_ATT.iloc[a,1]
        TravelTime_AB[int(a/2),1] = df_ATT.iloc[a,8]

        DelayAverage_AB[int(a/2),0] = df_ATT.iloc[a,1]
        DelayAverage_AB[int(a/2),1] = df_ATT.iloc[a,3]

        AverageSpeed_AB[int(a/2),0] = df_ATT.iloc[a,1]
        AverageSpeed_AB[int(a/2),1] = df_ATT.iloc[a,5]

        LatentDemand_AB[int(a/2),0] = df_ATT.iloc[a,1]
        LatentDemand_AB[int(a/2),1] = df_ATT.iloc[a,15]

        EffectiveDelay_AB[int(a/2),0] = df_ATT.iloc[a,1]
        EffectiveDelay_AB[int(a/2),1] = df_ATT.iloc[a,16]
```

```python
#DATA WITH NORMAL BUSES
TravelTime_B = np.zeros((int(lengthATT/2),2))
DelayAverage_B = np.zeros((int(lengthATT/2),2))
AverageSpeed_B = np.zeros((int(lengthATT/2),2))
LatentDemand_B = np.zeros((int(lengthATT/2),2))
EffectiveDelay_B = np.zeros((int(lengthATT/2),2))
#print(len(LatentDemand_B))
```

```python
#DATA WITH NORMAL BUSES
for a in range(0,lengthATT):

    if (a%2) == 0:
        
        TravelTime_B[int(a/2),0] = df_ATT.iloc[a,1]
        TravelTime_B[int(a/2),1] = df_ATT.iloc[a,8]

        DelayAverage_B[int(a/2),0] = df_ATT.iloc[a,1]
        DelayAverage_B[int(a/2),1] = df_ATT.iloc[a,3]

        AverageSpeed_B[int(a/2),0] = df_ATT.iloc[a,1]
        AverageSpeed_B[int(a/2),1] = df_ATT.iloc[a,5]

        LatentDemand_B[int(a/2),0] = df_ATT.iloc[a,1]
        LatentDemand_B[int(a/2),1] = df_ATT.iloc[a,15]

        EffectiveDelay_B[int(a/2),0] = df_ATT.iloc[a,1]
        EffectiveDelay_B[int(a/2),1] = df_ATT.iloc[a,16]
```

```python
print("TSTT")
print(np.subtract(TravelTime_B, TravelTime_AB))
print("average delay")
print(np.subtract(DelayAverage_B,DelayAverage_AB))
print("average speed")
print(np.subtract(AverageSpeed_B,AverageSpeed_AB))
print("latent demand")
print(np.subtract(LatentDemand_B,LatentDemand_AB))
print("effective delay")
print(np.subtract(LatentDemand_B,LatentDemand_AB))
```
