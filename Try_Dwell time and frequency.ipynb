```python
#!/usr/bin/env python
# coding: utf-8

print("This is for 0% CAV penetration and seed 1") #just for observing activity on console

import pandas as pd
import numpy as np
import time 
import threading
import math 
from time import sleep
from multiprocessing import Process
import os #Import os for creating a directory

t=time.time() #running time for code

#creating a dictionary for buses
#Insert the address of the fhz file 
print("importing vehicle ID information and creating dict for bus IDs")
######################################################
df = pd.read_csv(r'/mnt/amolikaStorage/Daniel/3x3FHZ.csv', delimiter = ';', names = ['Time','Link','Lane', 'VehNo','VehType', 'Line', 'DesSpeed'], dtype = {'Time' : float , 'Link' : float, 'Lane' : float, 'VehNo' : float, 'VehType' : float,'Line' : float, 'DesSpeed' : float }, skiprows =1, index_col=False)

Bus = df.loc[df['VehType'] == 300]  #Check which ID is for CABs and change here acording to that. 
Bus = Bus['VehNo']
Bus = Bus.to_csv((r"/mnt/amolikaStorage/Daniel/BUS.csv"))
Bus= pd.read_csv(r"/mnt/amolikaStorage/Daniel/BUS.csv", delimiter= ',', names = ['VehicleID'])
length_bus = len(Bus)
print("number of buses in the network ="+ str(length_bus))

#Import the link and lane information:
print("importig bus stop lane and link information")
#########################################################
df_link = pd.read_csv(r'/mnt/amolikaStorage/Daniel/Bus_stop_data.csv', delimiter = ',', index_col=False)
length_link = len(df_link)
print("number of links in the network"+str(length_link))

#PARAMETERS -> USER DEFINED
simulation_time_step=0.05 #simulation time step
threadblocks=70
minlinkV=1
maxlinkV=(350+1)
#READING THE FZP CONVERTED CSV FILE AND GETTING THE CORRESPONDING VALUES
print("Importing fzp file")
df = pd.read_csv('/mnt/amolikaStorage/Daniel/3x3FZP.csv', delimiter=';', names = ['time', 'no', 'link', 'lane', 'pos', 'poslat'], dtype={"time": float, "no": int, "link": int, "lane": int, "pos": float, "poslat": float },  skiprows =2, index_col=False)
print("Bingo! fzp file imported :)")
#This is the address of the input file.


def split_data(number1, number2): #number 1 and number 2 are the time 
    df_15= df.loc[(df['time'] > number1) & (df['time'] <= number2)]
    return df_15

# df_15 = df.loc[(df['Time'] >= 0) & (df['Time'] <= 15 minutes)]
# df_30 = df.loc[(df['Time'] > 15 minutes) & (df['Time'] <= 30 minutes)]
# df_45 = df.loc[(df['Time'] > 30 minutes) & (df['Time'] <= 45 minutes)]
# df_60 = df.loc[(df['Time'] > 45 minutes) & (df['Time'] <= 60 minutes)]

#number1=0
#number2=1800
def calc(number1, number2): #import  pandas dataframe as argument

    df = split_data(number1, number2)

    for p in range (0, length_link):

        link_number = df_link.iloc[p,0] #Input the link number here
        lane_number = df_link.iloc[p,1] #Input the lane number

        link_data = df.loc[df['link'] == link_number]
        link_lane_data = link_data.loc[link_data['lane'] == lane_number]

        link_inform = pd.DataFrame(np.zeros((length_bus, 7)), columns=['time','link','lane','bus ID','dwell time freq','dwell time','trajectory_length']) # creating an empty dataframe 

        for a in range (0, length_bus):
            trajectory = link_lane_data.loc[link_lane_data['no'] == Bus.iloc[a,0]]
            length_trajectory = len(trajectory)
            #print("link number ="+str(link_number)+ "lane number ="+ str(lane_number)+ "vehicle number ="+ str(Bus.iloc[a,0])+ "length of trajectory file"+ str(length_trajectory))
            count = 0
            if length_trajectory>1:
                print(Bus.iloc[a,0])
                print(trajectory)
                #count = 0
                
                for x in range (1, length_trajectory):
                    Speed = pd.DataFrame(np.zeros(((length_trajectory-1), 1)), columns=['Speed']) # creating an empty dataframe 
                    Speed.iloc[(x-1),0] = trajectory.iloc[x,4] - trajectory.iloc[(x-1),4]
                    print(Speed)
                    count = len(Speed.loc[Speed['Speed'] == 0])
                    
                print("dwell time number"+str(count))
            link_inform.iloc[a,0] = number2
            link_inform.iloc[a,1] = link_number
            link_inform.iloc[a,2] = lane_number
            link_inform.iloc[a,3] = Bus.iloc[a,0]
            link_inform.iloc[a,4] = count
            link_inform.iloc[a,5] = (count*simulation_time_step)
            link_inform.iloc[a,6] = length_trajectory
        link_inform.loc[(link_inform!=0).any(axis=1)]
        link_inform.to_csv((r"/mnt/amolikaStorage/Daniel/Results/" +"_Link_"+str(link_number) +"_time_"+ str(number2) +"_DT.csv"))
        
    print("Bazinga! link "+ str(link_number) + "is completed")
#calc(0,1200)
############

# 15 minutes would be 1800s, 
# 30 minutes would be 2700s, 
# 45 minutes would be 3600s; 
# 1 hour would be 4500s;

t1 = Process(target=calc, args= (0,1800))
print("Attaching process 1")
t2 = Process(target=calc, args= (1800,2700))
print("Attaching process 2")
t3 = Process(target=calc, args= (2700,3600))
print("Attaching process 3")
t4 = Process(target=calc, args= (3600,4500))
print("Attaching process 4")

t1.start()
print("Process 1 attached")
t2.start()
print("Process 2 attached")
t3.start()
print("Process 3 attached")
t4.start()
print("Process 4 attached")

t1.join()
print("Process 1 JOINING")
t2.join()
print("Process 2 JOINING")
t3.join()
print("Process 3 JOINING")
t4.join()
print("Process 4 JOINING")

print("done in : ", time.time()-t)
print("Bazinga! I have completed the job :D :D :D :D")
```
