fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()

# 15 Minutes

```python
fig, ax1 = plt.subplots(figsize=(10,8))


ax1.set_title("0-15 Minutes", fontweight = "bold", fontsize = 15)
#color = 'tab:red'
ax1.set_xlabel('% CAV Penetration', fontweight = "bold", fontsize = 13)
#ax1.set_ylabel('Number of Lane Change', color=color)
ax1.set_ylabel('Number of Lane Change', fontweight = "bold", fontsize = 13)
LN_CAB = ax1.plot(t, LN_CAB_1800, color="red", label = "CABs Network: Lane Change")
LN_Bus = ax1.plot(t, LN_Bus_1800, color="coral", label = "Bus Network: Lane Change")
#ax1.tick_params(axis='y', labelcolor=color)
ax1.tick_params(axis='y')
ax1.set_ylim(LaneChangeLower,LaneChangeHigh)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

#color = 'tab:blue'
#x2.set_ylabel('Dwell Time', color=color)
ax2.set_ylabel('Dwell Time' , fontweight = "bold", fontsize = 13)# we already handled the x-label with ax1
DL_CAB = ax2.plot(t, DL_CAB_1800, color="blue", label = "CABs Network: Dwell Time")
DL_Bus = ax2.plot(t, DL_Bus_1800, color="aqua", label = "Bus Network: Dwell Time")

#ax2.tick_params(axis='y', labelcolor=color)
ax2.tick_params(axis='y', size=20)
ax2.set_ylim(DwellTimeLower,DwellTimeHigh)


# added these three lines
lns = LN_CAB+LN_Bus+DL_CAB+DL_Bus
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc='best', bbox_to_anchor=(1.0, 0.5))


#ax1.legend(loc=0)
#ax2.legend(loc=4)

#ax1.legend()
#ax2.legend()

fig.tight_layout()  # otherwise the right y-label is slightly clipped

plt.show()
```
